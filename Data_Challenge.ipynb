Import Libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Load the Dataset

```
df = pd.read_csv('/content/omairSales.csv')
print(df)
```

Calculate average profits for each product

```
Portofits_Prod = df.groupby(['Prod'])['Sale'].mean() - df.groupby(['Prod'])['Cost'].mean()
Portofits = Portofits_Prod.mean()
print(Portofits_Prod ,Portofits )
```

Visulization data

```

Portofits_Prod.plot(kind='bar', color='skyblue')

# Titles
plt.title('Average Profit per Product')
plt.xlabel('Products')
plt.ylabel('Average Profit')
plt.axhline(y=Portofits, color='red', linestyle='--')

# Add profit values next to each bar
for index, value in enumerate(Portofits_Prod):
    plt.text(index, value - 300, f'{value:.1f}$', ha='center', color='black')

# Display the average profit value on the plot
plt.text(x=len(Portofits_Prod) - 3, y=Portofits + 30,
         s=f' {Portofits:.1f}$', color='red', fontsize=10, ha='center')
plt.show()
```
